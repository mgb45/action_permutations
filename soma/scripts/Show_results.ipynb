```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
iters_sink = np.load('iters_sink.npy')
iters_tcn = np.load('iters_tcn.npy')
iters_random = np.load('iters_random.npy')
```

```python
print(iters_sink.shape)
```

```python
plt.hist(iters_sink.ravel(),alpha=0.2)
plt.hist(iters_tcn.ravel(),alpha=0.2)
plt.hist(iters_random.ravel(),alpha=0.2)
# plt.plot(np.mean(iters_random,axis=1))
plt.show()
```

```python
plt.plot(np.mean(iters_sink,axis=0))
plt.plot(np.mean(iters_tcn,axis=0))
plt.plot(np.mean(iters_random,axis=0))
# plt.plot(np.mean(iters_random,axis=1))
plt.xlabel('puzzle')
plt.ylabel('Average iters')
plt.show()
```

```python
print('Average iterations rand: %2.2f +- %2.2f'%(np.mean(iters_random),np.std(iters_random)))
print('Average iterations tcn : %2.2f +- %2.2f'%(np.mean(iters_tcn),np.std(iters_tcn)))
print('Average iterations sink: %2.2f +- %2.2f'%(np.mean(iters_sink),np.std(iters_sink)))
```

```python
print('Random Success: %2.2f +- %2.2f'%(np.mean(np.sum((iters_random==0),axis=1)),np.std(np.sum((iters_random==0),axis=1))))
print('Sink Success: %2.2f +- %2.2f'%(np.mean(np.sum((iters_sink==0),axis=1)),np.std(np.sum((iters_sink==0),axis=1))))
print('TCN Success: %2.2f +- %2.2f'%(np.mean(np.sum((iters_tcn==0),axis=1)),np.std(np.sum((iters_tcn==0),axis=1))))
```
